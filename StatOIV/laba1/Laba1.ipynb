```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("data_resourse.csv")
```

```python
dataset
```

```python
coloums_target = ["Taxon Rank", "Start date day", "Start date month", "Start date year"]
filtered_dataset = dataset[coloums_target].copy()
filtered_dataset
```

```python
filtered_dataset["Taxon Rank"] = filtered_dataset["Taxon Rank"].replace({"species": 1, "family" : 2, "genus" : 3, "species hybrid" : 4})
filtered_dataset.dropna(inplace=True)
filtered_dataset.reset_index(drop=True, inplace=True)
```

```python
filtered_dataset
```

```python
filtered_dataset.to_csv("filtered_dataset.csv")
```
